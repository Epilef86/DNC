```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
wine_white = pd.read_csv("/content/winequality-white.csv", sep = ";")
wine_white.head()
```

```python
wine_red = pd.read_csv("/content/winequality-red.csv", sep = ";")
wine_red.head()
```

```python
wine_white["type"] = "white"
wine_white.head()
```

```python
wine_red["type"] = "red"
wine_red.head()
```

Juntando as duas tabelas por linha

```python
wines = pd.concat([wine_white, wine_red], axis = 0)
wines.head()
```

Avaliando vinho de acordo com a qualidade: 0 ruim e 1 é boa

```python
wines.loc[wines['quality'] <= 6, 'Avaliation'] = 0
wines.loc[wines['quality'] > 6, 'Avaliation'] = 1
 
wines.head(20)
```

Transformando variáveis categóricas em numerica

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

wines['type'] = le.fit_transform(wines.type.values)
wines
```

Verificar se tem dados nulos

```python
wines.info()
```

Remover dados  duplicados, pois quando é duplicado significa que toda linha é igual
---



```python
wines = wines.drop_duplicates()
wines.info()
```

Mundao a coluna Avaliation de float para int

```python
wines['Avaliation'] = wines['Avaliation'].astype(int)
wines.info()
```

Verificar se tem dados desbalanceados, ex: muito de um tipo e pouco de outro

```python
wines.Avaliation.value_counts()
```

Ou seja tem 4x mais de um tipo do que do outro, precisamos fazer um desbalanceamento. Vamos testar 03 formas existentes:

1ª smote --> cria dados que não são reais(sintéticos) para preencher o dataset com a classe minoritaria. O smote é usado para o dataset de treino e não para o dataset global, se não influenciará no dataset de teste, pois estaremos predizendo dados sintéticos. 

2ª near miss/randomUnderSample --> ela remove dados da classe majoritária, podemos aplicá-la antes de separar o dataset de teste do de treino

3ª é uma mistura das duas --> remover dados majoritários e criar dados sintéticos minoritários (sempre aplicando no treino, pois depois de treinado aplica no dado real)

y é a variável target

```python
y = wines['Avaliation']


x = wines.drop(columns={"Avaliation", "quality"}, axis=1) #pegando todas menos essas duas, tirei quality porque ela está ligada diretamente com a avaliação
```

Normalizar as variáveis pra que os valores fiquem o mais próximos possíveis

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)

x
```

Separar uma parte para treino e outra para teste

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

```

```python
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y, random_state = 0)
```

Aplicando o "smote" para balancear os dados de treino

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()
x_train, y_train = smote.fit_resample(x_train, y_train)
```

Usando a árvore de decisão para treinar o modelo

```python
from sklearn import tree
```

```python
arvore = tree.DecisionTreeClassifier()
```

```python
modelo = arvore.fit(x_train, y_train)
```

Fazer uma previsão para verificar se bate com modelo real

```python
y_pred = modelo.predict(x_test)
```

Imprimir um relatório com as métricas do modelo para validar se o modelo está bom

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, y_pred))
catbnm1 = classification_report(y_test, y_pred)
```

Observando a tabela acima verifica que o modelo não está bem equilibrado pois a precisão está muito alta para o "0" e baixa para o "1"

A árvore de decisão não foi muito bem ao se utilizar o balanceamento "smote". Vou testar o randomUnderSampler

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()
```

Aplicando o RandomUnderSampler diretamento no (x e y) antes de treinar, mais indicado

```python
from sklearn import tree

x = wines.drop(['Avaliation','quality'], axis = 1)
y =  wines['Avaliation']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)


from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()

x, y = rus.fit_resample(x,y)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y, random_state = 0)

model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)
# fazer as previsões em cima dos dados de teste 
y_pred = model.predict(x_test)
y_proba = model.predict_proba(x_test)

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test, y_pred))
catbnm1 = classification_report(y_test, y_pred)

print(pd.crosstab(y_test, y_pred, rownames = ['Real'], colnames = ['Predito'], margins = True))

y_test = y_test.to_frame()
```

Conclusão: o 2º modelo ele perfoma melhor, pois têm as métricas equilibradas entre a classe "0" e "1"
