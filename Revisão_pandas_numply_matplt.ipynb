```python
# transformar coluna data para o tipo datetime
df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
#Filtrar duas colunas com alguma condição e usando o E
df[(df['price']>=510000.0) & (df['bedrooms'] == 3)]
```

```python
#Filtrar duas colunas com alguma condição e usando o OU
df[(df['price']>=510000.0) | (df['bedrooms'] == 3)]
```

```python
#Removendo duplicatas

df.drop_duplicates(inplace=True)
```

```python
#Selecionar determinadas colunas

df_colunas = df[['bathrooms','sqft_lot','view']]
display(df_colunas)
```

```python
df.head()
```

```python
#Agrupar através de uma coluna(essa não tira a média) escolhida qual a média das demais colunas
media_zip = df.groupby(['zipcode']).agg('mean')
display(media_zip)
```

```python
#Agrupar através de uma coluna(essa não tira a média) escolhida qual a média de uma coluna específica
agg_media_zip = df.groupby(['zipcode'])['price'].agg('mean')
agg_media_zip
```

```python
#Se eu quiser agrupar por duas colunas a média de uma coluna específica
agg_zip_bedrooms = df.groupby(['zipcode','bedrooms'])['price'].agg('mean')
agg_zip_bedrooms
```

```python
#Agrupa uma coluna tendo uma outra com três métricas

grouped_zip = df.groupby('zipcode').agg({'price': ['mean', 'min', 'max']})
grouped_zip
```

```python
# IMPORTANTE

# MUITO USADA


#Através de uma coluna, quero agregar a média em duas colunas, aonde está mean, poderia ser soma(sum), count(contagem) ou qualquer operação

df.groupby('zipcode').agg(
    preco = ('price', 'mean'),
    acima_solo = ('sqft_above', 'mean')
)
```

```python
#Correlação--> quanto uma variável acaba sendo explicada por outra, ou seja, quando uma acontece, acontece outra também
#Tem que ver a teoria, quanto mais próximo de um mais forte a correlaçao
df.corr()
```
