```python
df = pd.read_csv('/content/kc_house_data.csv')
df.head()
```

```python
agg_zip = df.groupby(['bedrooms'])['price'].agg('mean').reset_index()
display(agg_zip)
```

```python
#Analisando graficamente por gr√°fico de barra
#Fun√ß√£o pra ajustar o tamanho da figura
fig = plt.figure(figsize=(20,5))
#gGr√°fico de barra, lembrando que o primeiro par√¢metro √© sempre o eixo x, que √© a coluna zipcode
plt.bar(agg_zip['bedrooms'], agg_zip['price'], color='green', width= 0.9)
#Legenda do eixo y
plt.ylabel("Pre√ßo")
#bold --> negrito
plt.title('Pre√ßo', loc='center', fontsize=18, fontweight = 'bold', color='blue')
```

```python
fig = plt.figure()
#gGr√°fico de barra deitado, lembrando que o primeiro par√¢metro √© sempre o eixo x, que √© a coluna zipcode
plt.barh(agg_zip['bedrooms'], agg_zip['price'])
#Legenda do eixo y
plt.ylabel("Pre√ßo")
#bold --> negrito
plt.title('Pre√ßo', loc='center', fontsize=18, fontweight = 'bold', color='blue')
plt.show()
```
