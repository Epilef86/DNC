```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/content/ee.xlsx')
```

```python
df
```

```python
df.info()
```

```python
Aumento_noturno_medio=df[df['Turno']=='noite']['%aumento'].mean()*100
```

```python
Aumento_matinal_medio=df[df['Turno']=='dia']['%aumento'].mean()*100
```

```python
print(Aumento_noturno_medio, Aumento_matinal_medio)
```

```python
x=[1,2]
plt.title('Produtividade média Noite x Dia')
plt.bar(x,height=[Aumento_noturno_medio,Aumento_matinal_medio])
plt.xticks(x, ('Noite','Dia'))
plt.show()
```

```python
plt.plot(df['% Geral'])   
plt.ylabel("% Geral")  
plt.title("Evolução do Progresso ao longo dos dias")  
plt.tick_params(
    axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)
plt.show()
```
